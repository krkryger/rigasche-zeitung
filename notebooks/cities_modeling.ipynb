```python
places['total_freq'] = df.placename.value_counts()[places.placename].values
```

```python
places = places.sort_values(by='total_freq', ascending=False)
```

```python
def annotate_cities(min_freq, input_string):
    
    cities = list(places.placename[places.total_freq >= 100].values)
    
    answers = []
    
    for city in cities:
        print('\n\n')
        print(city)
        choice = input(input_string)
        if choice != 'n':
            answers.append(1)
        else:
            answers.append(0)
            
    return answers
```

```python
german = annotate_cities(10000, 'Dominating German cultural presence? y/[N]')
```

```python
places['german_presence'] = german + [pd.NA]*(len(places) - len(german))
```

```python
russian = annotate_cities(100, 'Is located in the Russian Empire? y/[N]')
```

```python
places['russian_empire'] = russian + [pd.NA]*(len(places) - len(russian))
```

```python
capitals = annotate_cities(100, 'Is a sovereign capital? y/[N]')
```

```python
places['sovereign_capital'] = capitals + [pd.NA]*(len(places) - len(capitals))
```

```python
def population_estimate(t, city):
    
    if t in [1800, 1850, 1900]:
        return int(pop.loc[(pop.city == city) & (pop.year == t), 'inhabitants in 000-s'].values[0] * 1000)
    
    if t > 1800 and t < 1850:
        t_1 = 1800
        t_2 = 1850
        pop_t1 = int(pop.loc[(pop.city == city) & (pop.year == 1800), 'inhabitants in 000-s'].values[0] * 1000)
        pop_t2 = int(pop.loc[(pop.city == city) & (pop.year == 1850), 'inhabitants in 000-s'].values[0] * 1000)
    elif t > 1850 and t < 1900:
        t_1 = 1850
        t_2 = 1900
        pop_t1 = int(pop.loc[(pop.city == city) & (pop.year == 1850), 'inhabitants in 000-s'].values[0] * 1000)
        pop_t2 = int(pop.loc[(pop.city == city) & (pop.year == 1900), 'inhabitants in 000-s'].values[0] * 1000)
        
    pop_t = int((pop_t1**((t_2-t)/(t_2-t_1)))*(pop_t2**((t-t_1)/(t_2-t_1))))
        
    return pop_t
```
